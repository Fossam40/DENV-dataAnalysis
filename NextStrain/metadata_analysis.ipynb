```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
#import pydataset
import scipy.stats
```

```python
dataset = pd.read_csv("metadata_all.tsv", sep = '\t')
print(len(dataset))
print(len(dataset.drop_duplicates())) # No duplicates !
```

```python
dataset["host"].value_counts().plot(kind = 'bar')
plt.title("Frequency of hosts")
plt.ylabel("Host")
plt.xlabel("Frequency")
```

```python
dataset["region"].value_counts().plot(kind = 'bar')
plt.title("Frequency of regions")
plt.ylabel("Region")
plt.xlabel("Frequency")
```

```python
# Plot of most frequent countries
(dataset["country"].value_counts()).nlargest(20).plot(kind = 'bar')
plt.title("Frequency of (top 20 most frequent) countries")
plt.ylabel("Country")
plt.xlabel("Frequency")
```

```python
plt.hist(dataset["genome_coverage"], 40)
plt.title("Histogram of genome_coverage")
plt.ylabel("genome_coverage")
plt.xlabel("Frequency")
```

```python
plt.hist(dataset["length"], 40)
plt.title("Histogram of length")
plt.ylabel("length")
plt.xlabel("Frequency")
```

```python
dataset["serotype_genbank"].value_counts().plot(kind = 'bar')
plt.title("Histogram of serotype_genbank")
plt.ylabel("serotype_genbank")
plt.xlabel("Frequency")
```

```python
dataset["genotype"].value_counts().plot(kind = 'bar')
plt.title("Histogram of genotype")
plt.ylabel("genotype")
plt.xlabel("Frequency")
```
